```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import yahoo_fin.stock_info as si

# get list of Dow tickers
# OTHER TICKERS, SP500 does not work

dow_list = si.tickers_dow()[:10]
print(dow_list)
 
# Get data in the current column for each stock's valuation table
dow_stats = {}
cnt=0
```

```python
for ticker in dow_list:
    cnt+=1
    # print(ticker,cnt)
    temp = si.get_stats_valuation(ticker)
    # print(temp)
    temp = temp.iloc[:,:2]
    temp.columns = ["Attribute", "Recent"]
 
    dow_stats[ticker] = temp
```

```python
dataframe_list = []
for item,value in dow_stats.items():
    value['ticker'] = item
    dataframe_list.append(value)

stats_df= pd.concat(dataframe_list)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.startdataengineering.com/post/3-key-points-to-help-you-partition-late-arriving-events/"

response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Get the text under <div class="post">
post_text = soup.find("div", class_="post").text

# Get the text in H1 tag under <div class="post">
h1_text = soup.find("div", class_="post").h1.text

# Get URL in <div class="entry-navigation--right">
next_page_url = soup.find("div", class_="entry-navigation--right").find("a").get("href")

print(post_text)
print(h1_text)
print(next_page_url)
```
